## Connecting to the Database

```python
import sqlite3
conn = sqlite3.connect('data.sqlite')
c = conn.cursor()
```

## Display the names of all the employees in Boston.
Hint: join the employees and offices tables.

```python
c.execute("""
SELECT firstName, lastName 
FROM employees e JOIN offices o ON e.officeCode == o.officeCode
WHERE city == \'Boston\';
""").fetchall()
```

## Do any offices have no employees?
Hint: Combine the employees and offices tables and use a group by.

```python
c.execute("""
SELECT count(employeeNumber), officeCode
FROM employees
GROUP BY officeCode;
""").fetchall()
```

## Level Up: Display the names of each product each employee has sold

```python
c.execute("""
SELECT DISTINCT employeeNumber, productName
FROM (((products p JOIN orderdetails od on p.productcode == od.productcode) pod 
    JOIN orders o ON pod.orderNumber == o.orderNumber) podo
    JOIN customers c ON podo.customerNumber == c.customerNumber) podoc
    JOIN employees e ON podoc.salesRepEmployeeNumber == e.employeeNumber
ORDER BY employeeNumber;
""").fetchall()
```

## Level Up: Display the Number of Products each employee has sold

```python
c.execute("""
SELECT employeeNumber, COUNT(productName)
FROM (((products p JOIN orderdetails od on p.productcode == od.productcode) pod 
    JOIN orders o ON pod.orderNumber == o.orderNumber) podo
    JOIN customers c ON podo.customerNumber == c.customerNumber) podoc
    JOIN employees e ON podoc.salesRepEmployeeNumber == e.employeeNumber
GROUP BY employeeNumber
ORDER BY employeeNumber;
""").fetchall()
```
